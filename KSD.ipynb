## TRANSFER LEARNING ~ Kidney  Stones Detection

* We are going to carry out a Transfer Learning project, analyzing which hyperparameters are best suited for solving the problem.

```python
from ultralytics import YOLO
import torch
torch.backends.cudnn.enabled = False
```

```python
model = YOLO("yolov8n.pt")
```

* With the pre-trained model, we will train the model for a few epochs to estimate the best optimizer.

```python
model.train(data='D:/Datasets/archive/data.yaml', epochs=5, optimizer='SGD', device='0')
```

```python
model.train(data='D:/Datasets/archive/data.yaml', epochs=5, optimizer='Adam', device='0')
```

```python
model.train(data='D:/Datasets/archive/data.yaml', epochs=5, optimizer='RMSProp', device='0')
```

* It is concluded that the best optimizer is stochastic gradient descent (SGD), which is the default for the YOLOv8 network.
* Now we will train this model a bit more to further analyze its potential.

```python
model.train(data='D:/Datasets/archive/data.yaml', epochs=15, optimizer='SGD', device='0')
```

* Vemos que obtenemos buenos resultados, pero con capacidad de mejorar aún más por lo que aumentaremos el número de épocas y reduciremos el learning rate para tratar de mejorar el resultado.
* Reduciremos el tamaño del lote a 8 para reducir el consumo de memoria

```python
model.train(data='D:/Datasets/archive/data.yaml', epochs=50, optimizer='SGD', device='0', lr0=0.002, batch=8)
```

```python
model.val(split='test')
```
